## WHAT'S JUPYTER NOTEBOOK?

Jupyter notebook is a free and open source web application that integrates codes along with their outputs, markdown, visualization and equations (using LaTeX typesetting) in a single file that makes the presentation of the procedure and results of a project more intuitive. It is a web-based application (runs in a web browser).

It is also worth mentioning that notebooks promote collabrative works since they can be shared and edited by different people.

Jupyter supports 3 core programming languages and that's where it name comes from;
- Julia
- Python
- R

## COMPONENTS OF JUPYTER NOTEBOOK
1. __THE KERNEL:__ The notebook is run by the kernel. _'The kernel could be thought of as the heart of the notebook'_.  It is the program that executes the lines of code in a notebook. The default kernel programming language is Python.
2. __CELLS:__ The cells are the place where we write texts or lines of code. The are two kinds of cells;
    - _Code Cells:_ The content in this cell is interpreted by the current kernel's programming language. The output of the code is shown below the code cell when the cell is run.
    - _Markdown Cells:_ These cells allow us to write formatted text, insert images. Markdown is a lightweight markup language. 

For the purposes of this course/discussion, we will use google colab hence we skip the installation process for jupyter notebook.

## RUNNING CELLS


By default, new cells added to the notebook are _code cells._ 

To add a markdown cell, click on the _text_ button the shows up when you hover the cursor over the current cell as in the image below.
![Markdown Cell](./images/img3.png)

There are two ways of running the current cell (either _code cell_ or _markdown cell_);
- Using _'Shift + Enter':_ This executes the current cell and moves to the next cell. If there is no cell after the current cell, a new one will be created.
- Using _'Ctrl + Enter':_ This executes the current cell but the cursor remains in the current cell.

__NOTE:__ 
- Whenever you run a _code cell,_ the result is saved hence it can be used in other cells. It is advised to structure your code in a manner that initiates a variable before using it to avoid getting errors when you run the entire notebook after restarting the kernel.
- When you run a _code cell,_ there is a small square on the left side of the cell with a number in it. This number indicates the number of _code cells_ that has been run.

## KEYBOARD SHORTCUTS FOR NAVIGATING A NOTEBOOK

- __Y:__ Changes the cell type to _code cell._
- __M:__ Changes the cell type to _markdown cell._
- __DD:__ Deletes the current cell (It doesn't seem to work in google colab).
- __A:__ Adds a new cell above the existing one.
- __B:__ Adds a new cell below the existing one.
- __Z:__ Undo a deleted cell.
- __X:__ Cut a cell.
- __C:__ Copy a cell.
- __V:__ Paste a cell.

## LET'S NOW FOCUS ON MARKDOWN

Markdown is a lightweight markup language with just plain text. It was designed to be as 'easy-to-read' and 'easy-to-understand' as possible. Markdown is potentially converted to HTML.

### MARKDOWN SYNTAX
#### HEADERS
- #  _# - Header 1_
- ## _## - Header 2_
- ### _### - Header 3_
- #### _#### - Header 4_
- ##### _##### - Header 5_
- ###### _###### - Header 6_

#### EMPHASIS (BOLD, ITALIC)
- **Bold text** (\*\*Bold text\*\*)
- __Alternative Bold text__ (\_\_Aternative Bold Text\_\_)
- _Italic text_ (\_Italic text\_)
- ***Bold and Italic text*** (\*\*\*Bold and Italic text\*\*\*)
- ~~Strikethrough~~ (\~\~Strikethrough\~\~)

__NOTE:__ You can escape special characters using the backslash (\\). For instance, \\\# Header 1 is just read as a normal text and not intepreted as an H1 header.

- To create a paragraph, each line is separated by a blank line.
- To add a line break, add two or more empty spaces after the sentence.
- To make a blockquote, we use the greater than symbol. For instance,
> This is a blockquote (the sentence is prefixed with \>)
- We can nest blockquotes by using as many greater than signs as needed.
- For horizontal line, we use (\_\_\_), (\-\-\-), or (\*\*\*)

#### LISTS
To create a numbered list, you simply create a space insert a space between the number followed by a period and the sentence. For instance
1. I am one
2. I am two

To create an unordered list, you simply insert a space between the hyphen and the sentence. For instance,
- Unordered one
- Unordered two


#### ADDING IMAGES
To add an image, you can use one of the syntax below,  
- ![name of image](link to image)
- ![name of image](link to image "hover text") For an image which displays a text when you hover the mouse over it

#### ADDING LINKS (HYPERLINKS)
To add a hyperlink, you can use one of the syntax below,  
- [Name](hyperlink)

#### ADDING CODES
- **INLINE CODES:**
- \`print('Hello World')\`
- \`\`\`print('Hello World')\`\`\`

- **LANGUAGE SPECIFIC CODES**
- To write insert a code written in a specific language (say python), use the syntax below;
- \`\`\`language name  
            code block  
\`\`\`

Examples are as follows;

```javascript
function(){
    console.log('Hello World')
}
```
