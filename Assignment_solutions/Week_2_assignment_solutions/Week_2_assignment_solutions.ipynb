```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('sleep.csv')
df.info()
```

 Question 1 and 2. Consider the table below

- | Variables | Datatypes | Visualization method              |
  |-----------|-----------|-----------------------------------|
  | totSAS    | float     | Histogram/Box and Whisker plot    |
  | sex       | object    | Barplot/Piechart                  |
  | age       | float     | Histogram/Box and Whisker plot    |
  | edlevel   | object    | Barplot/Piechart                  |

- If a student does a plot of each variable, it is fine.
- The distribution for totSAS is normal(zero skewness).
- The distribution for age is skewed to the right.



```python
# Boxplot Visualization for totSAS
fig,ax = plt.subplots(figsize = (10,6))
df['totSAS'].plot(kind = 'box', vert = False, title = 'Distribution of totSAS')
plt.show()
```

```python
# Histogram visualization of totSAS
plt.figure(figsize=(10, 6))
plt.hist(df['totSAS'])
plt.xlabel('totSAS')
plt.ylabel('Frequency')
plt.show()
```

```python
# a barplot Visualization for sex
sex_counts = df['sex'].value_counts()
plt.figure(figsize=(10,6))
plt.bar(sex_counts.index, height=sex_counts.values, color =['red','yellow'])
plt.title('Distribution of Sex')
plt.xlabel('Sex')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pie chart of sex
plt.figure(figsize=(10,6))
plt.pie(sex_counts, labels=sex_counts.index,autopct='%1.1f%%')
plt.title('Gender')
plt.legend(sex_counts.index, loc='upper right')
plt.show()
```

```python
# Boxplot Visualization for age
fig,ax = plt.subplots(figsize = (10,6))
df['age'].plot(kind = 'box', vert = False, title = 'Distribution of Age')
plt.show()
```

```python
# Histogram visualization of age
plt.figure(figsize=(10, 6))
plt.hist(df['age'])
plt.xlabel('age')
plt.ylabel('Frequency')
plt.show()
```

```python
# a barplot Visualization for sex
edlevel_counts = df['edlevel'].value_counts()
plt.figure(figsize=(10,6))
bars = plt.bar(edlevel_counts.index, height=edlevel_counts.values, color =['red', 'blue','green','yellow','violet'])
plt.title('Distribution of edlevel')
plt.xlabel('edlevel')
plt.xticks(rotation=45)
plt.ylabel('Frequency')
# Add labels to the bars
for bar, label in zip(bars, edlevel_counts.index):
    bar.set_label(label)
plt.legend()
plt.show()
```

```python
# Pie chart of edlevel
plt.figure(figsize=(10,6))
plt.pie(edlevel_counts, labels=edlevel_counts.index,autopct='%1.1f%%')
plt.title('edlevel')
plt.legend(edlevel_counts.index, loc='lower left', bbox_to_anchor=(1, 0.5))
plt.show()
```
